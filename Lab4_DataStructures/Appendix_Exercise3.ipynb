```python
import array
from collections import deque

# 1. Initialize variables based on your student-specific parameters
# (Assume SEED_NUM and CONTROL_NUM are defined as per lab instructions)
# For this example, we use the sample values from your manual
BASE_VAL = 3 
VEC_DIM = 7
CONTROL_NUM = 4 # max(3, last_digit_of_id)

# 2. Initialize a memory-efficient Integer Array
# 'i' represents signed integers
signature_storage = array.array('i', [BASE_VAL, VEC_DIM, 100])

# 3. Initialize a sliding window using Deque with a maximum length
# The window size is determined by your CONTROL_NUM
window_size = CONTROL_NUM + 2
event_window = deque(maxlen=window_size)

# 4. Simulated Event Stream
event_stream = [101, 102, 103, 104, 105, 106, 107, 108]

# 5. Processing Loop
# This simulates adding events and maintaining only the most recent 'window_size'
for event in event_stream:
    event_window.append(event)
    # If the event matches a specific criteria (e.g., even number), 
    # store it in the long-term signature array
    if event % 2 == 0:
        signature_storage.append(event)

# 6. Outputs for Assessment Table
print(f"Final Integer Array (Signatures): {signature_storage}")
print(f"Final Sliding Window (Deque): {event_window}")
print(f"Array Memory Info: {signature_storage.buffer_info()}")
```
